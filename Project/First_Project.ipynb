```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('nelson_bay_max_temp.csv')
df.head(10)
```

```python
df.columns
```

```python
#clean up column names
def column_clean(column_name):
    new_column = []
    for i in column_name:
        i = i.lower().replace(' ', '_')
        new_column.append(i)
    return new_column
```

```python
df.columns = column_clean(df.columns)
df.columns
df.head()
```

```python
# Unnecessary to keep columns 'Product Code' and Bureau of Meterology station number
df = df.drop(df.columns[[0,1]], axis = 1)
df.head()
```

```python
# drop rows until 1st max_temp value
df = df.drop([df.index[1:198]])
df.head()
```
